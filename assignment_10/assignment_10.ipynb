```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
countries = pd.read_csv('data/countries_income.csv')
sns.set(rc={'figure.figsize': [10, 10]}, font_scale=1.2)
```

```python
countries.drop('2023', axis=1, inplace=True)
countries.head()
```

```python
sns.histplot(countries['RANK'], bins=49)
```

```python
sns.barplot(x='GROWTH', y='COUNTRIES', data=countries)
```

```python
sns.histplot(countries['GROWTH'])
```

```python
sns.jointplot(x='NATIONAL_INCOME', y='GROWTH_PER_CAPITA', data=countries, kind='scatter')
```

```python
sns.jointplot(x='NATIONAL_INCOME', y='GROWTH_PER_CAPITA', data=countries, kind='reg')
```

```python
sns.jointplot(x='NATIONAL_INCOME', y='GROWTH_PER_CAPITA', data=countries, kind='kde', shade=True)
```

```python
sns.heatmap(countries.corr(), annot=True)
```

```python
sns.countplot(y='CONTINENT', data=countries)
```

```python
sns.catplot(x="GROWTH_PER_CAPITA", y="COUNTRIES", kind="swarm", data=countries, height=6, aspect=1.3)
```

```python
c = sns.PairGrid(countries, height=3, aspect=1.3)
c.map_diag(sns.kdeplot, shade=True)
c.map_offdiag(sns.kdeplot, shade=True)
```

```python
sns.pairplot(countries, height=3, aspect=1.3)
```
